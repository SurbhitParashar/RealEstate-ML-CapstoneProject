```python
df=df[df['price']!='Price on Request']
```

```python
def convert_to_crore(value):
    value = str(value).strip().lower()

    if 'crore' in value:
        return float(value.replace('crore', '').strip())

    elif 'lac' in value or 'lakh' in value:
        num = (
            value.replace('lac', '')
                 .replace('lakh', '')
                 .strip()
        )
        return float(num) / 100

    return None  # or float(value) if needed
```

```python
df['price']=df['price'].apply(convert_to_crore)
```

```python
df['price_per_sqft']=df['price_per_sqft'].str.split().str[1].str.split('/').str[0].str.replace(",","").astype(float)
```
